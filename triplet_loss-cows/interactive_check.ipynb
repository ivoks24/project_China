```python
from similarity_check_2 import FeatureModel
import matplotlib.pyplot as plt
import numpy as np
```

```python
feature_model = FeatureModel('cowsResnet50.h5',
                             img_dir='cowsNose',
                             load_model=False,
                             input_size=(224, 224, 3))
```

```python
result = np.argmax(feature_model.get_all_features(img_dir='cowsNose'))
```

```python
result
```

```python
def pred(puth, threshold=None, save_unknown=False):

    newimg = plt.imread(puth)
    plt.imshow(newimg)
    plt.title(f'{puth}')
    plt.show()

    im_path = feature_model.get_most_similar(puth, threshold=threshold, save_unknown=save_unknown)

    pth = im_path[0][1].split('/')[1]
    print("real cow number: "+puth.split('/')[1]+"\nprediction: "+pth+"\ndistance: "+"%.4f"%im_path[0][0])
```

```python
def pred_f(puth, threshold=None, save_unknown=False):

    im_feature = feature_model.get_feature(puth)
    print(im_feature)
    
    newimg = plt.imread(puth)
    plt.imshow(newimg)
    plt.title(f'{puth}')
    plt.show()

    im_path = feature_model.get_most_similar(puth, threshold=threshold, save_unknown=save_unknown)
    pth = im_path[0][1].split('/')[1]
    
    print("cow number: "+pth+"\ndistance: "+"%.4f"%im_path[0][0])
```

```python
cows = {}
cowsList = feature_model.get_all_features('cowsNose')
count = len(cowsList[0])

for numCow in range(count):
    cows[cowsList[0][numCow].split('/')[1]] = cowsList[1][numCow]
    
print('\nCount cows: ' + str(count) + '\n')
for i in cows:
    print(i, ':', cows[i])
```
